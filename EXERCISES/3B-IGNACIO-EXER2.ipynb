**1. Creating a Data Set**

```python
import pandas as pd

data = {
    'student id': range(1, 16),
    'gender': ['Male','Male', 'Male','Male','Male','Female','Female','Female','Female','Female','Male','Male','Male','Male',''],
    'test_date': pd.date_range(start = '2024-02-14', periods= 15),
    'study_hours': [5,2,5,2,6,3,1,5,3,4,2,4,6,4,3],
    'test_score': [80,85,87,78,75,76,88,95,85,79,78,81,81,95,90],
    'attendance':['Present','Present','Present','Absent','Absent','Absent','Present','Present','Present','Absent','Absent','Absent','Present', 'Present','Present'],
}

df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index=False)
```

**2. Data Loading and Initial Examination**

```python
df = pd.read_csv('student_performance.csv')

print(df.head(16))
print(df.isnull().sum())
```

**3. Exploratory Data Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['test_score','study_hours']].describe()

#Histogram
plt.figure(figsize = (12,5))

plt.subplot(1,2,1)
sne.histplot(df['test_score'], bins=15, kde=True)
plt.title('Distribution of Test Score')

#Scatterplot
plt.subplot(1,2,2)
sne.histplot(df['study_hours'], bins = 15, kde=True)
plt.title('Distribution of Study Hours')

#Scatterplot
plt.figure(figsize=(5,5))
sne.scatterplot(data=df, x='study_hours', y='test_score')
plt.title('Scatter Plot of Study Hours vs. Test Scores')

```

**4. Data Visualization**

```python
#Gender Distribution
plt.figure(figsize=(6,4))
sne.countplot(data=df, x='gender')
plt.title('Gender Distribution')

#Test Score Over Time
plt.figure(figsize=(16,6))
sne.lineplot(data=df, x='test_date', y='test_score')
plt.title('Trend in Test Score Over Time')

correlation_matrix = df[['study_hours','test_score','attendance']].corr()
sne.heatmap(correlation_matrix, annot= True, cmap='coolwarm')
plt.title('Corelation Heatmap')
```
