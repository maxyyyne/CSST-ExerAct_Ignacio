**1. Import Libraries**

```python
import pandas as pd #Data Manipulation
import numpy as np #Mathematical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.linear_model import LinearRegression #Model Creation  that consists of Algorithm / Training Data / Data Modeling
```

**2. Data Loading**

```python
df = pd.read_csv('Lesson_4_student_grades_dataset.csv')
```

**3. Initial Data Analysis**

```python
stats = df.describe()
print(stats)

print(df.columns)
```

**4. Visualizing the Data (using scatterplot)**

```python
previous_exam_scores = df['Previous_Exam_Scores']
grades = df['Grades']

plt.figure(figsize=(12, 10))
plt.scatter(previous_exam_scores, grades, color='violet')

plt.title('Previous Exam Scores vs Grades')
plt.xlabel('Previous Exam Scores')
plt.ylabel('Grades')
plt.grid(True)
plt.show()
```

**5. Preparing Data for Modeling**

```python
X = df[['Previous_Exam_Scores']]
Y = df['Grades']
```

**6. Fitting the Linear Regression Model**

```python
model = LinearRegression()
model.fit(X, Y)
```

**7. Getting Model Coefficients**

```python
intercept = model.intercept_
slope = model.coef_[0]
print("Intercept:", intercept)
print("Slope:", slope)
```

**8. Visualizing the Linear Regression Line**

```python
plt.figure(figsize=(12, 10))
plt.scatter(previous_exam_scores, grades, color='violet')
plt.plot(X, model.predict(X), color='blue')
plt.title('Previous Exam Scores vs Grades with Linear Regression Line')
plt.xlabel('Previous Exam Scores')
plt.ylabel('Grades')
plt.grid(True)
plt.show()
```

**9. Making Predictions**

```python
new_previous_exam_scores = [[80], [85], [90]]
predicted_grades = model.predict(new_previous_exam_scores)
print("Predicted Grades for New Exam Scores:", predicted_grades)
```
