**1. Data Loading:**


*   Load the hardwareStore.csv data set into pandas DataFrame



```python
import pandas as pd #Data Manipulation
import matplotlib.pyplot as plt #Data Visualization
import seaborn as sne

df = pd.read_csv('hardwareStore.csv')
```

**2. Initial Data Analysis:**

*   Display basic statistics of the dataset, including mean, median, and standard deviation of  numeric columns
*   Identify the unique categories and their counts.



```python
stats = df.describe()
print(stats)

print(df.columns)
```

**3. Category Analysis:**

*   Create a bar chart to visualize the distribution of product across categories
*   Determine which category has the highest number of products.


```python

plt.figure(figsize=(10, 6))
df['CATEGORY_NAME'].value_counts().plot(kind='bar')
plt.title('Distribution of Products Across Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

highest_category = df['Category'].value_counts().idxmax()
print("Category with the highest number of products:", highest_category)
```

**4. Cost and Price Analysis:**


*   Calculate the average standard cost and average list price.
*   Identify the product with the highest standard cost and the one with the highest list price.





```python
average_standard_cost = df['STANDARD_COST'].mean()
average_list_price = df['LIST_PRICE'].mean()
print("Average Standard Cost:", average_standard_cost)
print("Average List Price:", average_list_price)

highest_standard_cost_product = df.loc[df['STANDARD_COST'].idxmax()]
highest_list_price_product = df.loc[df['LIST_PRICE'].idxmax()]
print("Product with the highest standard cost:", highest_standard_cost_product['PRODUCT_ID'])
print("Product with the highest list price:", highest_list_price_product['PRODUCT_ID'])
```

**5. Location Analysis:**


*  Determine the location (city or state) with the highest quantity of products in stock.



```python
location_with_highest_quantity = df.loc[df['QUANTITY'].idxmax()]['STATE']
print("Location with the highest quantity of products in stock:", location_with_highest_quantity)
```

**6. Insights and Reporting:**


*   Summarize your findings and insights from the data analysis.
*   Create a report or presentation that includes visualizations and explanation.



```python
highest_standard_cost_product = df.loc[df['STANDARD_COST'].idxmax()]
highest_list_price_product = df.loc[df['LIST_PRICE'].idxmax()]

location_with_highest_quantity = df.loc[df['QUANTITY'].idxmax(), 'STATE']

category_counts = df['CATEGORY_NAME'].value_counts()
highest_category = category_counts.idxmax()

avg_standard_cost = df['STANDARD_COST'].mean()
avg_list_price = df['LIST_PRICE'].mean()


print("SUMMARY OF STATISTICS:")
print("Average Standard Cost:", avg_standard_cost)
print("Average List Price:", avg_list_price)

print("\nCategory Analysis:")
print("Category with the highest number of products:", highest_category)

print("\nCost and Price Analysis:")
print("Product with the highest standard cost:", highest_standard_cost_product['PRODUCT_NAME'])
print("Product with the highest list price:", highest_list_price_product['PRODUCT_NAME'])

print("\nLocation Analysis:")
print("Location with the highest quantity of products in stock:", location_with_highest_quantity)

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
category_counts.plot(kind='bar')
plt.title('Distribution of Products Across Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)

plt.subplot(2, 1, 2)
data = {'Attribute': ['STANDARD_COST', 'LIST_PRICE'],
        'Average': [avg_standard_cost, avg_list_price]}
plt.table(cellText=[data[col] for col in data.keys()],
          colLabels=list(data.keys()), loc='center')
plt.axis('off')

plt.tight_layout()
plt.show()
```
