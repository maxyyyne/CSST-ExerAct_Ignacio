**1. Data Loading:**
  
•	Load the hardwareStore.csv dataset into a pandas DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('hardwareStore.csv')
```

**2. Initial Data Analysis**

•	Display basic statistics of the dataset, including mean, median, and standard deviation of numeric columns.

```python
statistics = data.describe( )

print(statistics)
```

•	Identify the unique categories and their counts.

```python
product_CATEGORY = data.groupby('CATEGORY_NAME')['CATEGORY_ID'].sum()

print(product_CATEGORY)
```

**3. Category Analysis:**

•	Create a bar chart to visualize the distribution of products across categories.

```python
product_CATEGORY = data.groupby('CATEGORY_NAME')['CATEGORY_ID'].sum()

product_CATEGORY.plot(kind = 'bar')
plt.xlabel('CATEGORY_NAME')
plt.ylabel('CATEGORY_ID')
plt.title('CATEGORY BY COUNT')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()


product_CATEGORY = data.groupby('CATEGORY_NAME')['DESCRIPTION'].count()

product_CATEGORY.plot(kind = 'bar')
plt.xlabel('CATEGORY_NAME')
plt.ylabel('DESCRIPTION')
plt.title('CATEGORY BY COUNT OF DESCRITION')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

•	Determine which category has the highest number of products.

```python
highest_number_of_product = product_CATEGORY.idxmax()
print(highest_number_of_product)
```

**4. Cost and Price Analysis:**

•	Calculate the average standard cost and average list price.

```python
average_standard_cost = data['STANDARD_COST'].mean
average_list_price = data['LIST_PRICE'].mean

print(average_list_price)
print(average_standard_cost)
```

•	Identify the product with the highest standard cost and the one with the highest list price.

```python
product_STANDARD_COST = data.groupby('CATEGORY_NAME')['STANDARD_COST'].sum()
product_LIST_PRICE = data.groupby('CATEGORY_NAME')['LIST_PRICE'].sum()

product_STANDARD_COSTING = data.groupby('STANDARD_COST')['CATEGORY_NAME'].count()
product_LIST_PRICING = data.groupby('LIST_PRICE')['CATEGORY_NAME'].count()

highest_number_of_standard_costing = product_STANDARD_COSTING.idxmax()
highest_number_of_pricing = product_LIST_PRICING.idxmax()

highest_number_of_standard_cost = product_STANDARD_COST.idxmax()
highest_number_of_price_list = product_LIST_PRICE.idxmax()


print(highest_number_of_standard_cost)
print(highest_number_of_price_list)
```

**5. Location Analysis:**

•	Determine the location (city or state) with the highest quantity of products in stock.

```python
product_QUANTITY_CITY = data.groupby('CITY')['QUANTITY'].count()
product_QUANTITY_STATE = data.groupby('STATE')['QUANTITY'].count()

highest_number_of_quantity_per_city = product_QUANTITY_CITY.idxmax()
highest_number_of_quantity_per_state = product_QUANTITY_STATE.idxmax()


print(highest_number_of_quantity_per_city)
print(highest_number_of_quantity_per_state)
```
