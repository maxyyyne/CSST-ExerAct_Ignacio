***1. Import Libraries:***

```python
import pandas as pd #data manipulation
import numpy as np #mathematical operation
import matplotlib.pyplot as plt #data visulalization
from sklearn.linear_model import LinearRegression #data modeling/data training
```

***2. Reading CSV file:***

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
```

***3. Assigning Variables***

```python
independent = data['Hours_of_Study']
dependent = data['Grades']
```

***4. Visualizing the Data***

```python
plt.scatter(independent, dependent)
plt.xlabel('Independent Varialble(independent)')
plt.ylabel('Dependent Varialble(dependent)')
plt.title('Scatter Plot of Data')
plt.show()
```

***5. Preparing Data for Modeling:***

```python
independent = data[['Hours_of_Study']]
dependent = data[['Grades']]
```

***6. Fitting the Linear Regression Model:***

```python
model = LinearRegression()
model.fit(independent,dependent)
```

***7. Getting Model Coefficients:***

```python
slope = model.coef_[0]
intercept = model.intercept_
print('Slope: ',slope)

print('Intercept: ',intercept)
```

***8. Visualizing the Linear Regression:***

```python
plt.scatter(independent, dependent)
plt.plot(independent,model.predict(independent), color = 'skyblue')
plt.xlabel('Independent Variable(independent)')
plt.xlabel('Dependent Variable(dependent)')
plt.title('Linear Regression')
plt.show()
```
