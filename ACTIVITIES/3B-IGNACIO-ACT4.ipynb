# ***1. Import Libraries***

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# ***2. Data Preperation***

```python
data = pd.read_csv('titanic_data.csv')
statistics = data.describe()
print(statistics)
```

```python
data.head()
```

# ***Correlation between Survived and other variables***

```python
corr_class_survive = data.groupby('Pclass')['Survived'].count()
corr_class_survive.plot(kind = 'bar')
plt.xlabel('Pclass')
plt.ylabel('Survived')
plt.title(' Pclass & Survived ')
plt.xticks(rotation = 0)
plt.tight_layout()
plt.show()

corr_sex_survive = data.groupby('Sex')['Survived'].count()
corr_sex_survive.plot(kind = 'bar')
plt.xlabel('Sex')
plt.ylabel('Survived')
plt.title('Sex & Survived')
plt.xticks(rotation = 0)
plt.tight_layout()
plt.show()

corr_survive_survive = data.groupby('Survived')['Survived'].count()
corr_survive_survive.plot(kind = 'bar')
plt.xlabel('Survived')
plt.ylabel('Survived')
plt.title('Survivability')
plt.xticks(rotation = 0)
plt.tight_layout()
plt.show()

corr_sex_sex = data.groupby('Sex')['Sex'].count()
corr_sex_sex.plot(kind = 'bar')
plt.xlabel('Sex')
plt.ylabel('Sex')
plt.title('Sex & Sex')
plt.xticks(rotation = 0)
plt.tight_layout()
plt.show()

corr_age_survive = data.groupby('Age')['Survived'].count()
corr_age_survive.plot(kind = 'line')
plt.xlabel('Age')
plt.ylabel('Survived')
plt.title('Age & Survived')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()


corr_sibsp_survive = data.groupby('SibSp')['Survived'].count()
corr_sibsp_survive.plot(kind = 'bar')
plt.xlabel('SibSp')
plt.ylabel('Survived')
plt.title('SibSp & Survived')
plt.xticks(rotation = 0)
plt.tight_layout()
plt.show()


corr_embarked_survive = data.groupby('Embarked')['Survived'].count()
corr_embarked_survive.plot(kind = 'bar')
plt.xlabel('Embarked')
plt.ylabel('Survived')
plt.title('Embarked & Survived')
plt.xticks(rotation = 0)
plt.tight_layout()
plt.show()


print(corr_class_survive)
print(corr_sex_survive)
print(corr_sex_sex)
print(corr_age_survive)
print(corr_sibsp_survive)
print(corr_embarked_survive)
print(corr_survive_survive)
```
